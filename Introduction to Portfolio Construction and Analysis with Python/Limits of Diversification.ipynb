## Limits of Diversification

```python
ind = erk.get_ind_returns()
ind_nfirms = erk.get_ind_nfirms()
ind_size = erk.get_ind_size()
```

The market cap of a particular industry is the sum of total value of the included companies present in the industry. Let's calculate the market cap of specific industries which would be the total number of companies in the industry times the average cap of the industry. Since we are multiplying 2 datasets of the same size element-wise, the resulting ind_mkcap  dataset should also have the same shape.

```python
ind_mkcap = ind_nfirms * ind_size
ind_mkcap.shape
```

Now, let's calculate the total yearly marketcap which would be the sum of marketcaps of individual industries.

```python
total_mkcap = ind_mkcap.sum(axis="columns")
total_mkcap.shape
```

```python
total_mkcap.plot(figsize=(10,5))
```

We can see that the trend of total market cap of all the industries since 1929. 

Let's get the ratio of market cap in each industry for the specified years in the dataset. That can be obtained by dividing the market caps on each industry by the total market cap in that year. 

```python
ind_capweight = ind_mkcap.divide(total_mkcap, axis="rows")
ind_capweight.head()
```

From  the result above we find out that in July 1926, the percent of market cap in Food industry was 5.8%. As you might expect, if we sum the cap weights for all the industries, it should give us a value of 1.0

```python
ind_capweight["1926"].sum(axis="columns")
```

We can plot the trend of individual cap weights for any 2 industries. for eg: Steel and fin

```python
ind_capweight[["Fin", "Steel"]].plot()
```

The above graph clearly states the increasing contribution of Finance industry to the overall market cap across years. Conversely, the market cap contribution of Steel industry has decreased.

We can find the yearly market returns obtained by multiplying the ind_capweight times the returns.

```python
total_market_returns = (ind_capweight * ind).sum(axis="columns")
total_market_returns.head()
```

```python
total_market_returns.plot()
```

We want to take the returns and compound them to get the market indices.

```python
total_market_index = erk.max_drawdown(total_market_returns).Wealth
total_market_index.plot(figsize=(10,5), title="Total Market Capitalisation Index 1926-2018")
```

We can calculate the exact same thing, only by giving a rolling window of 36 months. So now instead of calculating the returns for each month, it will calculate the mean for 36 months window altogether.

We will see that the plot for the rolling window starts a bit later than the other one. That's because, for the first data point to be plotted, it needs atleast 36 months.

```python
total_market_index.plot(figsize=(12,6))
total_market_index.rolling(window=36).mean().plot(figsize=(10,5))
```

Now we'll look at the returns of the index and not the index itself. We want to compound the results for each window of the rolling plot. This can be done using .aggregate(). For each window we will call the annualize_rets function.

```python
tmi_tr36rets = total_market_returns.rolling(window=36).aggregate(erk.annualized_rets, 12)
tmi_tr36rets.plot(figsize=(10,5), title="Tr 36 months returns", legend=True)
total_market_returns.plot(label="Returns", legend=True)
```

Since we have got the returns of trailing 36 months, we can find out the level of correlation between the industries over this period.

## Rolling Correlation

```python
rolling_corr = ind.rolling(window=36).corr()
rolling_corr.tail()
```

The problem we are facing here is that there are 2 indexes to the correlation data. The data tells us the measure of correlation between any 2 industries in a given month. For eg: the correlation between the Retail and Food industries in Dec 2018 was 51.7%

```python
rolling_corr.index.names = ["Date", "Industry"]
rolling_corr.tail()
```

Now, let's find the mean of the correlations between the industries for all the months. We can get the mean by grouping the data based on date.

```python
ind_36corr = rolling_corr.groupby(level="Date").apply(lambda cormat: cormat.values.mean())
ind_36corr.plot(figsize=(10,5))
```

Let's superimpose the rolling monthly returns and correlation on one graph. Since both lie on the different scale, we will need to a secondary y axis scale.

```python
tmi_tr36rets.plot(title="Tr36 Mo Rets", figsize=(10,5), legend=True)
ind_36corr.plot(title="Tr36 Mo Corr", figsize=(10,5), legend=True, secondary_y=True)
```

We can clearly see that as the returns are falling, the correlation between the industries is actually increasing. This would be the time when you  would want the correlation between the stocks in your portfolio to reduce, but instead it spikes upwards. Thats's the reason diversification is not advantageous always.
