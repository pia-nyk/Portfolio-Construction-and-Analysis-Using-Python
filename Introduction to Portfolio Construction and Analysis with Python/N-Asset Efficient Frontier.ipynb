# N-Asset Efficient Frontier

Calculating the N Asset Efficient Frontier will be very similar to calculating the 2 Asset Efficient Frontier except for the allocation of weights to the assets. For 2 asset, assigning one weight would automatically assign the other weight, this won't work in case of N assets.

```python
%load_ext autoreload
%autoreload 2
import pandas as pd
import edhec_risk_kit as erk
from scipy.optimize import minimize
```

```python
ind = erk.get_ind_returns()
er = erk.annualized_rets(ind["1996":"2000"], 12)
cov = ind["1996":"2000"].cov()
```

We are going to find the weights of a portfolio which gives us a certain return at lowest volatility.

```python
l = ["Smoke", "Fin", "Games", "Coal"]
erk.plot_ef(25, er[l], cov.loc[l,l])
```
